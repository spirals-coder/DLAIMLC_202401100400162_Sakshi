```python
#Given Data
X = [75, 65, 80]
W = [4, 3, 5]
b = -1000
eta = 0.01
y = 1

# Net Input (z)
z = 0
for i in range(len(X)):
    z = z + (W[i] * X[i])
z = z + b

print("Net input z =", z)

#  Apply Activation Function
if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Predicted Output =", y_pred)

#Calculate Error
error = y - y_pred
print("Error =", error)

#Weight Update
if error != 0:
    for i in range(len(W)):
        W[i] = W[i] + (eta * error * X[i])

#  Bias Update
b = b + (eta * error)

# Print Updated Values
print("Updated Weights =", W)
print("Updated Bias =", b)

# Re-check after learning
z_new = 0
for i in range(len(X)):
    z_new = z_new + (W[i] * X[i])
z_new = z_new + b

print("New z =", z_new)

if z_new >= 0:
    print("Final Prediction = Pass")
else:
    print("Final Prediction = Fail")
```
