```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_sentence = ["I", "love", "AI"]
output_sentence = ["Je", "aime", "IA"]
```

```python
attention_weights = np.array([
    [0.2, 0.6, 0.2],   # Je attends mostly to "love"
    [0.7, 0.2, 0.1],   # aime attends to "I"
    [0.1, 0.3, 0.6]    # IA attends to "AI"
])
```

```python
plt.figure(figsize=(6,5))

plt.imshow(attention_weights, cmap='viridis')

# Label axes
plt.xticks(range(len(input_sentence)), input_sentence)
plt.yticks(range(len(output_sentence)), output_sentence)

plt.xlabel("Input Sentence")
plt.ylabel("Output Sentence")
plt.title("Attention Heatmap")

plt.colorbar()
plt.show()
```
